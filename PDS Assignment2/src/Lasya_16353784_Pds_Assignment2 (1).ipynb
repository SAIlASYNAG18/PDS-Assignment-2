## A) Looking for the missing values

```python
data.isna().sum()
```

```python
d2= pd.DataFrame(data)
```

## B) Removing the units from Mileage, Engine, Power and New_Price and converting them into float to keep in numerical values

```python
d2['Mileage'] = d2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

d2['Engine'] = d2['Engine'].str.replace(' CC', '').astype(float)

d2['Power'] = d2['Power'].str.extract('(\d+\.\d+)').astype(float)

d2['New_Price'] = d2['New_Price'].str.extract('(\d+\.\d+)').astype(float)

d2.head()
```

## (A) Filling in missing values im Mileage, Engine, Power,Seats with median

```python
d2['Mileage'].fillna(d2['Mileage'].median(),inplace=True)
d2['Engine'].fillna(d2['Engine'].median(),inplace=True)
d2['Power'].fillna(d2['Power'].median(),inplace=True)

d2['Seats'].fillna(d2['Seats'].mode()[0],inplace=True)
d2=d2.dropna()
d2.head()
```

```python
output_csv_file = 'preprocessed_data.csv'
d2.to_csv(output_csv_file, index=False)
```

Determine which columns in this job have missing values: Mileage (2), Engine (36), Power (36), Seats (38) and New Price (5032). Since there are more missing values in this instance, the New Price column was removed from the data frame. As the number of missing information for mileage, engine, power, and seats decreases assigned a "median" value to them. When compared to imputation using mean, the robust strategy of imputing missing values using the median is less susceptible to outliers and works well for a variety of data distributions.

## C) Changing the categorical variables to numerical value

```python
d1=pd.read_csv("/content/preprocessed_data.csv")
```

```python
#To work with our categorical variables, we are converting them into numerical data using one-hot encoding.
d1['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
d1['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
d1['Owner_Type'].replace({'First': 1, 'Second': 2, 'Third': 3, 'Fourth & Above': 4}, inplace=True)
```
